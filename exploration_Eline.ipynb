```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the data
df = pd.read_csv('data/raw/correlation_jngdc.csv')
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
correlation_target = np.abs(df.corr().target) > 0.9
```

```python
df_corr = pd.DataFrame(df.corr())
corr_list = df_corr.mask(np.triu(np.ones(df_corr.shape, dtype=np.bool_))).unstack()
print('Correlation between two variables')
print(corr_list[(corr_list.notnull()) & (abs(corr_list) > 0.5)].sort_values())
```

```python
plt.scatter(df['99'], df['target'])
plt.xlabel('variable 99')
plt.ylabel('target')
plt.show()
```
